```julia
m=1
x=range(0,1,step=0.01/m)
initX=x
```

```julia
function newfunc(x)
    return tanh.((x ./ 4) ./ (sqrt.(1 .- x .^ 2)))
end;

y=newfunc.(x)
using Plots

scatter(y,zeros(length(y)),ylim=[-1,1])
#scatter(x,zeros(length(y)),ylim=[-1,1])
```

```julia
#calculate dx~/dx
function der_grid(X)

    aux=[]

    for i in 1:length(X)
        aux = vcat(aux,Der(X,i,initX))
    end
    
    #dergrid_func = Spline1D(X[ori:L-3],aux,k=4, bc="extrapolate")

    return aux#dergrid_func
end

function Der(y,i,X) #dx tilde/dx
    
    if i==1 # left boundary TEM1
        z = (-27*y[i]+58*y[i+1]-56*y[i+2]+36*y[i+3]-13*y[i+4]+2*y[i+5])/(12*(X[i+1]-X[i]))
    elseif i==2 # left boundary TEM2
        z = (-2*y[i-1]-15*y[i]+28*y[i+1]-16*y[i+2]+6*y[i+3]-y[i+4])/(12*(X[i+1]-X[i]))
    elseif i==length(y) # left boundary TEM1
        z = -(-27*y[i]+58*y[i-1]-56*y[i-2]+36*y[i-3]-13*y[i-4]+2*y[i-5])/(12*(X[i]-X[i-1]))
    elseif i==length(y)-1 # left boundary TEM2
        z = (-2*y[i+1]-15*y[i]+28*y[i-1]-16*y[i-2]+6*y[i-3]-y[i-4])/(12*(X[i+1]-X[i]))
    else # central
        z = (-y[i+2]+8*y[i+1]-8*y[i-1]+y[i-2])/(12*(X[i+1]-X[i]))
    end
    
    return z
end
```

```julia
function integrate(f,x0,x2)
    x1 = (x0 .+ x2) ./ 2
    dx = (x2 .- x0) ./2
    
    z = dx ./ 3 .* (f.(x0) .+ 4 .* f.(x1) .+ f.(x2))
    return z
end
```

```julia
function func(x)
    return 100 ./ (1 .+ exp.(.- x.^2 .+ 5))
end
```

```julia
y=integrate(func,0,x)
```

```julia
using Plots
scatter(y,zeros(length(y)),ylim=[-1,1])
```
